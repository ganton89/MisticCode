ğŸ¯ **DescripciÃ³n breve:**  
Un clÃ¡sico reinventado. Este proyecto implementa en Python el famoso juego  
**â€œPiedra, Papel, Tijeraâ€**, ampliado con las divertidas variantes  
**Lagarto** y **Spock**, popularizadas en *The Big Bang Theory*.  

El jugador se enfrenta al ordenador en un duelo de estrategia, azar y risas,  
con mensajes explicativos y un marcador final.  

---

```python
# Opciones: definimos las opciones del juego (Piedra, Papel, Tijera, Lagarto, Spock) con una lista
opciones = ["piedra", "papel", "tijera", "lagarto", "spock"]

# Reglas: diccionario de sets. Cada jugada gana a los elementos de su set.
reglas = {
    "tijera": {"papel", "lagarto"},
    "papel": {"piedra", "spock"},
    "piedra": {"tijera", "lagarto"},
    "lagarto": {"spock", "papel"},
    "spock": {"tijera", "piedra"}
}

# Motivos (ganador, perdedor) -> explicaciÃ³n
motivos = {
    ("tijera", "papel"):   "Tijera âœ‚ï¸ corta papel ğŸ“„",
    ("papel", "piedra"):   "Papel ğŸ“„ cubre piedra ğŸª¨",
    ("piedra", "lagarto"): "Piedra ğŸª¨ aplasta lagarto ğŸ¦",
    ("lagarto", "spock"):  "Lagarto ğŸ¦ envenena Spock ğŸ––",
    ("spock", "tijera"):   "Spock ğŸ–– rompe tijera âœ‚ï¸",
    ("tijera", "lagarto"): "Tijera âœ‚ï¸ decapita lagarto ğŸ¦",
    ("lagarto", "papel"):  "Lagarto ğŸ¦ come papel ğŸ“„",
    ("papel", "spock"):    "Papel ğŸ“„ refuta Spock ğŸ––",
    ("spock", "piedra"):   "Spock ğŸ–– vaporiza piedra ğŸª¨",
    ("piedra", "tijera"):  "Piedra ğŸª¨ aplasta tijera âœ‚ï¸"
}

import random

def jugada_maquina():
    return random.choice(opciones)

def jugada_jugador():
    while True:
        # Manejamos las posibles interrupciones/errores, recogiendo este bloque de cÃ³digo en un try/except
        try:
            entrada = input("Elige una opciÃ³n: piedra, papel, tijera, lagarto o spock: ").strip().lower()
        except (KeyboardInterrupt, EOFError):
            print("\nEntrada interrumpida. Por favor, elige de nuevo.")
            continue

        # Como es posible que el jugador introduzca "Spok", en lugar de "Spock", "normalizamos" esta entrada para que coincida
        if entrada == "spok":
            entrada = "spock"

        if entrada in opciones:
            return entrada

        else:
            print("âš ï¸ OpciÃ³n no vÃ¡lida. IntÃ©ntalo de nuevo.")

# Muestra el motivo adecuado segÃºn la pareja (ganador, perdedor)
def mensaje(op1, op2):
    if (op1, op2) in motivos:
        print(motivos[(op1, op2)])
    elif (op2, op1) in motivos:
        print(motivos[(op2, op1)])
    else:
        print("CombinaciÃ³n no vÃ¡lida")

# Marcadores
puntuacion_usuario = 0
empate = 0
puntuacion_cpu = 0

# Rondas
ronda = 1

print("=== Piedra ğŸª¨, Papel ğŸ“„, Tijera âœ‚ï¸, Lagarto ğŸ¦, Spock ğŸ–– ===")
print("Modo: 5 rondas\n")
print("Â¿Listo?\n")

while ronda <= 5:
    usuario = jugada_jugador()
    cpu = jugada_maquina()

    print(f"\nRonda: {ronda}")
    print("----------------------------------------------------")
    print('Jugador elige:', usuario)
    print('Ordenador perverso elige:', cpu)

    if usuario == cpu:
        print('Â¡Vaya! Ha habido un empate')
        empate += 1

    elif cpu in reglas[usuario]:
        # Gana el jugador -> motivo(jugador, cpu)
        print('Â¡Ganaste!')
        mensaje(usuario, cpu)
        puntuacion_usuario += 1

    else:
        # Gana la mÃ¡quina -> motivo(cpu, jugador)
        print('Â¡Ups! Perdiste')
        mensaje(cpu, usuario)
        puntuacion_cpu += 1

    print('----------------------------------------------------')
    ronda += 1

# Resultados finales
print("\n=== Resultado final ===")
print("El jugador ha ganado:", puntuacion_usuario, "vez/veces")
print("El ordenador perverso ha ganado:", puntuacion_cpu, "vez/veces")
print("Total empates:", empate, "vez/veces")

if puntuacion_usuario > puntuacion_cpu:
    print('ğŸ† Â¡Enhorabuena!Â¡Has ganado la partida!')
elif puntuacion_usuario < puntuacion_cpu:
    print('ğŸ’» Â¡Oh,oh! El ordenador perverso ganÃ³ la partida.')
    print('\n Â¡Hasta luego, Maricarmen!')
else:
    print('ğŸ¤ Vaya... un empate, Â¿quieres la revancha?')
```
