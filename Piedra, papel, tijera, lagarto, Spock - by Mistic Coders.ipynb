**"Piedra, papel, tijera, lagarto, Spock"**

---

🎯 **Descripción breve:**  
Un clásico reinventado. Este proyecto implementa en Python el famoso juego  
**“Piedra, Papel, Tijera”**, ampliado con las divertidas variantes  
**Lagarto** y **Spock**, popularizadas en *The Big Bang Theory*.  

El jugador se enfrenta al ordenador en un duelo de estrategia, azar y risas,  
con mensajes explicativos y un marcador final.  

---

```python
# Opciones: definimos las opciones del juego (Piedra, Papel, Tijera, Lagarto, Spock) con una lista
opciones = ["piedra", "papel", "tijera", "lagarto", "spock"]

# Reglas: diccionario de sets. Cada jugada gana a los elementos de su set.
reglas = {
    "tijera": {"papel", "lagarto"},
    "papel": {"piedra", "spock"},
    "piedra": {"tijera", "lagarto"},
    "lagarto": {"spock", "papel"},
    "spock": {"tijera", "piedra"}
}

# Motivos (ganador, perdedor) -> explicación
motivos = {
    ("tijera", "papel"):   "Tijera ✂️ corta papel 📄",
    ("papel", "piedra"):   "Papel 📄 cubre piedra 🪨",
    ("piedra", "lagarto"): "Piedra 🪨 aplasta lagarto 🦎",
    ("lagarto", "spock"):  "Lagarto 🦎 envenena Spock 🖖",
    ("spock", "tijera"):   "Spock 🖖 rompe tijera ✂️",
    ("tijera", "lagarto"): "Tijera ✂️ decapita lagarto 🦎",
    ("lagarto", "papel"):  "Lagarto 🦎 come papel 📄",
    ("papel", "spock"):    "Papel 📄 refuta Spock 🖖",
    ("spock", "piedra"):   "Spock 🖖 vaporiza piedra 🪨",
    ("piedra", "tijera"):  "Piedra 🪨 aplasta tijera ✂️"
}

import random

def jugada_maquina():
    return random.choice(opciones)

def jugada_jugador():
    while True:
        # Manejamos las posibles interrupciones/errores, recogiendo este bloque de código en un try/except
        try:
            entrada = input("Elige una opción: piedra, papel, tijera, lagarto o spock: ").strip().lower()
        except (KeyboardInterrupt, EOFError):
            print("\nEntrada interrumpida. Por favor, elige de nuevo.")
            continue

        # Como es posible que el jugador introduzca "Spok", en lugar de "Spock", "normalizamos" esta entrada para que coincida
        if entrada == "spok":
            entrada = "spock"

        if entrada in opciones:
            return entrada

        else:
            print("⚠️ Opción no válida. Inténtalo de nuevo.")

# Muestra el motivo adecuado según la pareja (ganador, perdedor)
def mensaje(op1, op2):
    if (op1, op2) in motivos:
        print(motivos[(op1, op2)])
    elif (op2, op1) in motivos:
        print(motivos[(op2, op1)])
    else:
        print("Combinación no válida")

# Marcadores
puntuacion_usuario = 0
empate = 0
puntuacion_cpu = 0

# Rondas
ronda = 1

print("=== Piedra 🪨, Papel 📄, Tijera ✂️, Lagarto 🦎, Spock 🖖 ===")
print("Modo: 5 rondas\n")
print("¿Listo?\n")

while ronda <= 5:
    usuario = jugada_jugador()
    cpu = jugada_maquina()

    print(f"\nRonda: {ronda}")
    print("----------------------------------------------------")
    print('Jugador elige:', usuario)
    print('Ordenador perverso elige:', cpu)

    if usuario == cpu:
        print('¡Vaya! Ha habido un empate')
        empate += 1

    elif cpu in reglas[usuario]:
        # Gana el jugador -> motivo(jugador, cpu)
        print('¡Ganaste!')
        mensaje(usuario, cpu)
        puntuacion_usuario += 1

    else:
        # Gana la máquina -> motivo(cpu, jugador)
        print('¡Ups! Perdiste')
        mensaje(cpu, usuario)
        puntuacion_cpu += 1

    print('----------------------------------------------------')
    ronda += 1

# Resultados finales
print("\n=== Resultado final ===")
print("El jugador ha ganado:", puntuacion_usuario, "vez/veces")
print("El ordenador perverso ha ganado:", puntuacion_cpu, "vez/veces")
print("Total empates:", empate, "vez/veces")

if puntuacion_usuario > puntuacion_cpu:
    print('🏆 ¡Enhorabuena!¡Has ganado la partida!')
elif puntuacion_usuario < puntuacion_cpu:
    print('💻 ¡Oh,oh! El ordenador perverso ganó la partida.')
    print('\n ¡Hasta luego, Maricarmen!')
else:
    print('🤝 Vaya... un empate, ¿quieres la revancha?')
```
